# 4.3.1 심층 신경망(DNN)

심층 신경망(DNN)은 입력층과 출력층 사이에 다수의 은닉층을 포함하는 인공 신경망이다.

다수의 은닉층을 포함하여 별도의 트릭 없이 다양한 비선형적 관계를 학습할 수 있다. 

그러나 학습을 위한 많은 연산량이 필요하고 과하게 학습하여 실제 데이터에 대해 오차가 증가하는 과적합, 기울기 값의 소실 문제 등의 문제점이 발생할 수 있다. 

→ 드롭아웃, 렐루 함수, 배치 정규화 등의 기법을 적용하면서 이러한 문제를 해결하고 심층 기계 학습(deep learning)의 핵심 모델로 활용되고 있다.

드롭 아웃: 실제로 훈련 정확도를 희생시키면서 테스트 정확도를 높여 과적 합을 방지하는 것을 목표로하는 정규화의 한 형태

렐루 함수: 활성화함수로 시그모이드의 기울기 소실 문제 해결 

배치 정규화: 학습 과정에서 각 배치 단위 별로 데이터가 다양한 분포를 가지더라도 각 배치별로 평균과 분산을 이용해 정규화하는 것

# 4.3.2 합성곱 신경망 (CNN)

합성곱 신경망(CNN)은 합성곱층과 풀링층을 포함하는 이미지 처리 성능에 좋은 인공 신경망 알고리즘이다.

영상 및 사진이 포함된 이미지 데이터에서 객체를 탐색하거나 객체 위치를 찾아내는데 유용한 신경망이다.

특히 이미지에서 객체, 얼굴, 장면을 인식하기 위해 패턴을 찾는 데 유용하다.

## 합성곱 연산 구성

합성곱 신경망은 합성곱 연산, 풀링 연산, 활성화 함수로 구성되어 있다.

일반적 심층신경망(DNN)은 이미지 데이터를 일반 데이터처럼 처리하기 위해 1차원으로 변환한다. 하지만, 1차원으로 변환된 이미지 데이터는 본래 이미지의 특징을 잘 담아내지 못한다.

따라서, 이미지 본연의 형태인 2차원을 1차원으로 바꾸지 않으면서 이미지의 특징을 추출할 수 있는 합성곱 연산을 신경망에 적용시켰다.

### 합성곱 연산: 이미지 픽셀의 고유값 그리고 필터 안의 가중치를 Element Wise하여 Feature map의 Pixel 하나에 입력하는 과정을 의미한다. 

*Element Wise: 대응하는 행렬의 요소들의 곱을 의미하는 연산과정

### 풀링 연산: feature map의 크기를 줄이는 역할을 한다.

## 기존 신경망(DNN)과 차이점

- 각 층의 입출력을 항상 유지한다.


- 이미지의 공간 정보를 유지하면서 인접 이미지와 차이가 있는 특징을 효과적으로 인식한다.


- 복수 필터로 이미지의 특징을 추출하고 학습한다.


- 추출한 이미지의 특징을 모으고 강화하는 풀링층이 있다.


- 필터(커널)를 공유 파라미터로 사용하기 때문에 일반 인공 신경망과 비교하여 학습 파라미터가 매우 적다.

# 4.3.3 순환 신경망(RNN)

순환 신경망(RNN)은 시계열 데이터(음악, 영상 등) 같은 시간 흐름에 따라 변화하는 데이터를 학습하기 위한 인공 신경망이다.

순환 신경망의 '순환'은 자기 자신을 참조한다는 것으로, 현재 결과가 이전 결과와 연관이 있다는 의미이다.

여러 개의 데이터가 순서대로 있을 때, 앞서 입력받은 데이터를 잠시 기억해놓고 기억된 데이터가 얼마나 중요한지 판단하고 별도의 가중치를 주어 다음 데이터로 넘어간다.

## 순환 신경망 특징

- 시간성을 가진 데이터가 많다.


- 시간성 정보를 이용하여 데이터의 특징을 잘 다룬다.


- 시간에 따라 내용이 변하므로 데이터는 동적이고, 길이는 가변적이다.


### 순환 신경망은 기울기 소멸 문제로 학습이 제대로 되지 않는 문제가 있다. 

#### 기울기 소멸문제: 은닉층을 많이 거칠수록 전달되는 오차가 크게 줄어들어 학습이 되지 않는 현상, 기울기가 거의 0으로 소멸되어 버리면 네트워크의 학습은 매우 느려지고 학습이 다 이루어지지 않은 상태에서 멈춘다.


→ 이를 해결하고자 메모리 개념을 도입한 LSTM이 순환 신경망에서 많이 사용된다. 

# 4.3.4 제한된 볼츠만 머신(RBM)

볼츠만 머신은 가시층과 은닉층으로 구성된 모델이다.

이 모델에서 가시층은 은닉층과만 연결되는데 (가시층과 가시층, 은닉층과 은닉층 상에 연결은 없는) 이것이 제한된 볼츠만 머신이다.

첫번째 층은 우리가 볼 수 있는 층인 가시층, 혹은 데이터가 입력되는 입력층이고 두번째 층은 특징값이 학습되는 은닉층이다.

입력층의 노드는 데이터를 입력받으면 입력받은 데이터를 은닉층에 얼마나 전달할 것인지를 확률에 따라 결정한다.

즉, 확률에 따라 입력을 전달할지 혹은 전달하지 않을지를 결정한다.

각 입력 노드는 데이터의 저수준 특징값 혹은 속성을 받아온다. 

예를 들어, 흑백 이미지가 입력 데이터라면 각 입력 노드는 이미지의 픽셀 값에 해당한다. 

입력으로 x를 넣으면 그 은닉층의 노드에 있는 가중치 w와 곱해진다. 그리고 bias(b)를 더한 값 w * x + b를 활성함수f()에 넣은 것이 노드의 출력이 된다.

## 제한된 볼츠만 머신의 특징

- 차원 감소, 분류, 선형 회귀 분석, 협업 필터링, 특성 값 학습, 주제 모델링에 사용된다.


- 기울기 소멸 문제를 해결하기 위해 사전 학습 용도로 활용 가능하다.


- 심층 신뢰 신경망(DBN)의 요소로 활용된다.

# 4.3.5 심층 신뢰 신경망(DBN)

심층 신뢰 신경망(DBN)은 입력층과 은닉층으로 구성된 제한된 볼츠만 머신을 블록처럼 여러 층으로 쌓은 형태로 연결된 신경망이다.

사전 훈련된 제한된 볼츠만 머신을 층층이 쌓아 올린 구조로, 레이블이 없는 데이터에 대한 비지도 학습이 가능하다. 

부분적인 이미지에서 전체를 연상하는 일반화와 추상화 과정을 구현할 때 사용하면 유용하다.


## 심층 신뢰 신경망 학습 절차

1. 가시층과 은닉층 1에 제한된 볼츠만 머신을 사전 훈련한다


2. 첫 번째 층 입력 데이터와 파라미터를 고정하여 두 번째 층 제한된 볼츠만 머신을 사전 훈련한다.


3. 원하는 층 개수만큼 제한된 볼츠만 머신을 쌓아 올려 전체 DBN을 완성한다.

## 심층 신뢰 신경망 학습 특징

- 순차적으로 심층 신뢰 신경망을 학습시켜 가면서 계층적 구조를 생성한다.


- 비지도 학습으로 학습한다.


- 위로 올라갈수록 추상적 특성을 추출한다.


- 학습된 가중치를 다층 퍼셉트론의 가중치 초깃값으로 사용한다.
